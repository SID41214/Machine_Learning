```python
import pandas.io.data as web
```

```python
df = web.DataReader('AAPL', 'google', '2016/1/1', '2017/1/1')
```

```python
df.head()
```

```python
%matplotlib inline
df.plot(y='Close', color="Green")
```

```python
df.plot.bar(y='Volume')
```
